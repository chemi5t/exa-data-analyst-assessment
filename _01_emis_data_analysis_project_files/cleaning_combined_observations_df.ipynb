# Cleaning of combined_observation.csv

```python
# Path to combined_observation.csv file
combined_observation_file_path = os.path.join(r'..\data', 'combined_observation.csv')

# Extract combined_observation.csv
combined_observation_df = data_extractor.extract_from_csv(combined_observation_file_path)
```

I know the comparator coloumn (columne 2) has object dtype from pdDataFrame.info(). I have seen the values of '>', '>=', '<=' and '>'.

Now going to filter and look at these rows.

```python
# # Show the rows where "comparator" isnotnull
combined_observation_df_isnull_mask = combined_observation_df['comparator'].isnull()
combined_observation_df_isnotnull_filtered = combined_observation_df[~combined_observation_df_isnull_mask]
display(combined_observation_df_isnotnull_filtered.info())
display(combined_observation_df_isnotnull_filtered)
```

```python
combined_observation_df['comparator'].unique()
```

```python
# Cast column to string
combined_observation_df['comparator'] = combined_observation_df['comparator'].astype(str)

# Replace 'nan' strings with actual NaN values
combined_observation_df.loc[combined_observation_df['comparator'] == 'nan', 'comparator'] = pd.NA

# Cast column to categorical type
combined_observation_df['comparator'] = combined_observation_df['comparator'].astype('category')

# Verify the changes
print(combined_observation_df['comparator'].unique(), "\n")
print(combined_observation_df['comparator'].dtype)
```

```python
def clean_column_data(selected_table_df: pd.DataFrame, column_name: str) -> pd.DataFrame:

        # Cast the column to string
        selected_table_df[column_name] = selected_table_df[column_name].astype(str)

        # Replace 'nan' strings with actual NaN values
        selected_table_df.loc[selected_table_df[column_name] == 'nan', column_name] = pd.NA

        # Convert the column to categorical type
        selected_table_df[column_name] = selected_table_df[column_name].astype('category')

        return selected_table_df
```

```python
combined_observation_df = clean_column_data(combined_observation_df, 'comparator')
display(combined_observation_df.info())
```
